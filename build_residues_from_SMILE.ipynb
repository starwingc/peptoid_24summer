```python
peptoid_path = 'peptoid/'

# Iterate over each column
for col_name, col_data in clean_smile_df.items():
    pep_savepath = os.path.join(peptoid_path, col_name)
    os.makedirs(pep_savepath, exist_ok=True)
    print(pep_savepath)
    
    for index, residue in col_data.items():
        if pd.isna(residue):
            continue
        else:
            command = f"obabel -:'{residue}' -omol2 --gen3d -O {pep_savepath}/{index}.mol2"
            process = subprocess.Popen(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, shell=True, text=True)
            stdout, stderr = process.communicate()
        
            print(command)
            print("Standard Output:")
            print(stdout)
            print("Standard Error:")
            print(stderr)
    
```
